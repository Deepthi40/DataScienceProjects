**SCRAPING THE MOBILE DATA FROM FLIPKART**
- Importing Necessary libraries to the notebook
-finding the Right URL of needed products
- finding the necessary html tags from url
- scraped multiple pages using loops
- created dataframe using pandas
- converted dataframe to **CSV** file.

```python
from bs4 import BeautifulSoup
import lxml
import requests
import datetime
import time
import pandas as pd
```

```python
name=[]
price=[]
rating=[]
reviews=[]
desc=[]
page=0
for pages in range(0,11):
  page=pages+1
  url=f"https://www.flipkart.com/search?q=mobile+phone&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_5_6_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_5_6_na_na_na&as-pos=5&as-type=[RECENT,Trending]&suggestionId=mobile+phone&requestId=6e4b57d9-96dd-47a7-b8f1-e8a94df79cc9&page={page}"
  headers= {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"}
  page=requests.get(url,headers=headers)
  soup=BeautifulSoup(page.text,"html.parser")
  #print(url)
  product_name=soup.find("div",class_="_1YokD2 _3Mn1Gg").find_all("div",class_=["_4rR01T","_30jeq3 _1_WHN1","_3LWZlK","gUuXy-","fMghEO"])
  name.extend([product.get_text() for product in product_name][0::5])
  rating.extend([product.get_text() for product in product_name][2::5])
  price.extend([product.get_text() for product in product_name][4::5])
  reviews.extend([product.get_text() for product in product_name][1::5])
  desc.extend([product.get_text() for product in product_name][3::5])
   

```

```python
df=pd.DataFrame({"prod_name":name,"prod_price":price,"star_rating":rating,"prod_rr":reviews,"prod_desc":desc})
df
```

```python
df.to_csv("mobile_data.csv",index=False,encoding='utf-8')
```
